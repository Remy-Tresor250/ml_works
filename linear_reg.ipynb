```python
import matplotlib.pyplot as plt
from scipy import stats

x=[5,7,8,8,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]
slope, intercept, r, p, std_err =stats.linregress(x,y)
print(slope)
print(intercept)
print(r)
print(p)
def my_func(x):
    return slope * x +intercept +std_err 
```

```python
speed = my_func(1)
print(speed)
```

```python
mymodel = list(map(my_func,x))
print(mymodel)
```

```python
plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```
